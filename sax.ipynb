```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('starbucks.csv')
```

```python
data.head(5)
```

```python
data.groupby(['Beverage','Beverage_prep'])['Total Fat (g)'].aggregate('max')
```

```python
## THE WHITE CHOCOLATE MOCHA WHICH WAS PREPARED WITH SOYMILK HAD THE MAXIMUM FAT
##MEASURED IN GRAMS.
```

```python
data.groupby(['Beverage_category','Beverage', "Beverage_prep"]).max()['Total Carbohydrates (g) '].sort_values(ascending=False)[:10].plot(kind='barh', color='orchid')
plt.xlabel('Total cabohydrates')
plt.ylabel('Top 10 drinks with High Cabohydrates content')
```
